```python
def hl(d):
    df = pd.DataFrame(columns=d.columns, index=d.index)
    #df.loc[d['WORD_RES'].ne(d['WORD_COR'])] = 'background: yellow'
    df.loc[d['WORD_RES'].ne(d['WORD_COR']), ['WORD_RES', 'WORD_COR']] = 'background: red'
    return df
```

```python
result = df_all.style.apply(hl, axis=None)
result
```
