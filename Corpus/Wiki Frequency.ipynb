```python
import pandas as pd
from collections import Counter
import collections
```

```python
colnames = ["text"]
df = pd.read_csv("sah_wikipedia_2021_100K.txt", sep="\t", encoding="utf-8", header=None, names=colnames)
df.head()
```

```python
#df = df.drop("index", axis=1)
```

```python
df["text"] = df["text"].str.replace("[^\w\s]", " ", regex=True)
```

```python
df["text"] = df["text"].str.replace("  ", " ")
```

```python
df["text"] = df["text"].str.lower()
```

```python
df["text"] = df["text"].str.replace("\n", "")
```

```python
df["text"] = df["text"].str.replace("\d*", "", regex=True)
```

```python
wiki_texts = " ".join(df["text"])
```

```python
dct = Counter(wiki_texts.split(" "))
```

```python
res = collections.OrderedDict(sorted(dct.items()))
res
```

```python
df_counter = pd.DataFrame.from_dict(res, orient='index').reset_index()
df_counter.to_csv("count_sah_words.tsv", sep="\t", index=False)
```
