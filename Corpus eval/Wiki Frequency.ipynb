```python
import pandas as pd
from collections import Counter
#import collections
```

```python
colnames = ["text"]
df = pd.read_csv("sah_wikipedia_2021_100K.txt", sep="\t", encoding="utf-8", header=None, names=colnames)
df.head()
```

```python
#df = df.drop("index", axis=1)
```

```python
df["text"] = df["text"].str.replace("[^\w\s]", " ", regex=True)
```

```python
df["text"] = df["text"].str.lower()
```

```python
df["text"] = df["text"].str.replace("\n", "")
```

```python
df["text"] = df["text"].str.replace("\d*", "", regex=True)
```

```python
df["text"] = df["text"].str.replace("\s+", " ")
df["text"] = df["text"].str.strip()
```

```python
wiki_texts = " ".join(df["text"])
wiki_texts = wiki_texts.split(" ")
```

## Removing non yakut words in corpus

```python
sahalpha = ["ҥ", "ҕ", "ө", "һ", "ү"]
```

```python
a = ord('а')
```

```python
rus = "".join([chr(i) for i in range(a, a + 6)] + [chr(a + 33)] + [chr(i) for i in range(a + 6, a + 32)])
rus
```

```python
for i in rus:
    sahalpha.append(i)
sahalpha[0:10]    
```

```python
test_vocab = []

for word in wiki_texts:
    for letter in word:
        if letter in sahalpha:
            #print(word)
            
            test_vocab.append(word)
            break
            
```

```python
dct = Counter(test_vocab)
#dct
```

```python
from operator import itemgetter
res = sorted(dct.items(), key=itemgetter(1), reverse=True)
res
```

```python
df = pd.DataFrame.from_dict(res)
df.columns = ["Word", "Count"]
```

```python
df.head()
```

```python
## Delete single letters
```

```python
df = df[df['Word'].str.len() > 1]
```

```python
print("Full corpus length : " + str(len(df)))
```

```python
df = df[df['Count'] >= 15]
```

```python
df.to_csv("counter_sah_wiki.tsv", sep="\t", index=False)
```
