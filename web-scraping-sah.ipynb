```python
import re
import requests
import newspaper
from bs4 import BeautifulSoup
from collections import Counter
```

```python
kyym_paper = newspaper.build('https://kyym.ru/', memoize_articles=False)
edersaas_paper = newspaper.build('https://edersaas.ru/', memoize_articles=False)
sakha_ysia_paper = newspaper.build('https://sakha.ysia.ru/', memoize_articles=False)
sakhaparliament_paper = newspaper.build('https://www.sakhaparliament.ru/all-tags/dokuuskaj-kuorat', memoize_articles=False)
```

```python
article_texts = []
```

```python
for index in range(len(kyym_paper.articles)):
    #print(index)
    first_article = kyym_paper.articles[index]
    first_article.download()
    first_article.html
    first_article.parse()

    print(first_article.text)
    article_texts.append(first_article.text)
```

```python
for index in range(len(sakha_ysia_paper.articles)):
    #print(index)
    first_article = sakha_ysia_paper.articles[index]
    first_article.download()
    first_article.html
    first_article.parse()

    print(first_article.text)
    article_texts.append(first_article.text)
    
    #print("#############")
```

```python
with open ("sah_news.txt", "w", encoding="utf-8") as w:
    for line in article_texts:
        if line != "":
            w.write(line)
```

```python
with open ("sah_news_clean.txt", "r", encoding="utf-8") as f:
    f = f.read()
```

```python
word_list = []
```

```python
f = re.sub("\n", "", f)
f = re.sub("[^\w\s]", "", f)
f = re.sub("\d*", "", f)
f = f.lower()
spl = f.split()
spl
```

```python
for word in spl:
    if len(word) > 1:
        word_list.append(word)
```

```python
len(Counter(word_list))
```

```python
count = Counter({k: c for k, c in Counter(word_list).items() if c >= 5})
```

```python
from operator import itemgetter
res = sorted(count.items(), key=itemgetter(1), reverse=True)
res
```

```python
len(res)
```

```python
nouns = []
verbs = []
adj = []
adv = []
part = []
conj = []
pron = []
num = []
```

```python
word_lst = list(count.keys())
```

```python
for word in word_lst:
    r1 = requests.get(f"https://sakhatyla.ru/translate?q={word}")
    coverpage = r1.content
    soup1 = BeautifulSoup(coverpage, 'html5lib')
    coverpage_news = soup1.find_all('em')
    
    for i in coverpage_news:
        if "v." in i:   
            res = word, i
            if res not in verbs:
                verbs.append(res)
        elif "n." in i:   
            res = word, i
            if res not in nouns:
                nouns.append(res)
            
        elif "a." in i:   
            res = word, i
            if res not in adj:
                adj.append(res)
                
        elif "conj." in i:  
            res = word, i
            if res not in conj:
                conj.append(res)
                
        elif "adv." in i:  
            res = word, i
            if res not in adv:
                adv.append(res)
                
        elif "int." in i or "мест." in i:  
            res = word, i
            if res not in pron:
                pron.append(res)           
                
        elif "частица" in i:   
            res = word, i
            if res not in part:
                part.append(res)
                
        elif "числ." in i:   
            res = word, i
            if res not in num:
                num.append(res)        
                   
        else:
            res = word, i
            print(res)
```
